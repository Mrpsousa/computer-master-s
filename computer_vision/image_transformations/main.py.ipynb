# Plot da Transformada de cor

```python
import skimage
# Split
img = resize(imread('cat.5000.jpg'), (500, 500))

red_c_mul = [1,0,0]
green_c_mul = [0,1,0]
blue_c_mul = [0,0,1]
shift = [3,3,3]

#Ligth intensity change
ligth_int_chg = (red_c_mul * img) + (green_c_mul * img) + (blue_c_mul * img)

#Ligth intensity shift
ligth_int_sft = img + shift

ligth_int_cng_plus_sft = ligth_int_chg + shift

img2 = skimage.img_as_uint(img)

red_channel = ligth_int_cng_plus_sft[:, :, 0]
green_channel = ligth_int_cng_plus_sft[:, :, 1]
blue_channel = ligth_int_cng_plus_sft[:, :, 2]

new_red = (red_channel - np.mean(red_channel))/np.std(red_channel)
new_green = (green_channel - np.mean(green_channel))/np.std(green_channel)
new_blue = (blue_channel - np.mean(blue_channel))/np.std(blue_channel)

img2[:, :, 0] = new_red
img2[:, :, 1] =new_green
img2[:, :, 2] =new_blue

fig, axs = plt.subplots(2,2, figsize=(10,8))

cax_00 = axs[0,0].imshow(img2)
axs[0,0].set_title('Original IMG - Ligth intensity change + shift - Transformada de cor')
axs[0,0].xaxis.set_major_formatter(plt.NullFormatter())  # kill xlabels
axs[0,0].yaxis.set_major_formatter(plt.NullFormatter())  # kill ylabels

cax_01 = axs[0,1].imshow(red_channel, cmap='Reds')
fig.colorbar(cax_01, ax=axs[0,1])
axs[0,1].set_title('')
axs[0,1].xaxis.set_major_formatter(plt.NullFormatter())
axs[0,1].yaxis.set_major_formatter(plt.NullFormatter())

cax_10 = axs[1,0].imshow(green_channel, cmap='Greens')
fig.colorbar(cax_10, ax=axs[1,0])
axs[1,0].set_title('Green Channel of Ligth intensity change + shift')
axs[1,0].xaxis.set_major_formatter(plt.NullFormatter())
axs[1,0].yaxis.set_major_formatter(plt.NullFormatter())

cax_11 = axs[1,1].imshow(blue_channel, cmap='Blues')
fig.colorbar(cax_11, ax=axs[1,1])
axs[1,1].set_title('Blue Channel of Ligth intensity shift')
axs[1,1].xaxis.set_major_formatter(plt.NullFormatter())
axs[1,1].yaxis.set_major_formatter(plt.NullFormatter())
plt.show()

# Plot histograms
fig, axs = plt.subplots(3, sharex=True, sharey=True, figsize=(9,7))

axs[0].hist(new_red.ravel(), bins=10)
axs[0].set_title('Red channel of Ligth intensity change + shift - Transformada de cor')
axs[1].hist(new_green.ravel(), bins=10)
axs[1].set_title('Green channel of Ligth intensity change + shift - Transformada de cor')
axs[2].hist(new_blue.ravel(), bins=10)
axs[2].set_title('Blue channel of Ligth intensity change + shift - Transformada de cor')

plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt
import skimage

img = resize(imread('cat.5000.jpg'), (500, 500))
img2 = skimage.img_as_uint(img)


red_channel = img[:, :, 0]
green_channel = img[:, :, 1]
blue_channel = img[:, :, 2]

new_red = (red_channel - np.mean(red_channel))/np.std(red_channel)
new_green = (green_channel - np.mean(green_channel))/np.std(green_channel)
new_blue = (blue_channel - np.mean(blue_channel))/np.std(blue_channel)

img2[:, :, 0] = new_red
img2[:, :, 1] = new_green
img2[:, :, 2] = new_blue


for i, col in enumerate(['b', 'g', 'r']):
    hist = cv2.calcHist([img2], [i], None, [256], [0, 256])
    plt.plot(hist, color = col)
    plt.xlim([-500, 500])
    

plt.show()
```
